### Tidy Data
https://pandas.pydata.org/pandas-docs/stable/user_guide/reshaping.html

```python
df = df_kosis.melt(id_vars="시군구별")
df.head()
```

```python
df["시군구별"].unique()
```

```python
df[df["시군구별"]=='시군구별'].head()
```

```python
# Remove useless columns
df[df["시군구별"]!='시군구별'].head()
```

```python
print(df.shape)
df = df[df["시군구별"]!='시군구별'].copy()
print(df.shape)
```

```python
df.head()
```

```python
df["variable"]
# for eaxample, 1997 - year
#               01 - month
#               1 or 2 - male or female
```

```python
# split Variable by '.'
df["variable"].str.split('.', expand=True)
```

```python
df['year'] = df['variable'].str.split('.', expand=True)[0]
df.head()
```

```python
df['month'] = df['variable'].str.split('.', expand=True)[1]
df.head()
```

```python
df['gender'] = df['variable'].str.split('.', expand=True)[2]
df.tail()
```

```python
df['gender'].unique()
```

```python
# fill 0(total) in gender None value
df['gender'] = df['gender'].fillna('0')
df.head(10)
```

```python
# But I want to change word not number in gender columns,
df['gender'] = df['gender'].replace('0', 'total').replace('1', 'male').replace('2', 'female')
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
df['gender'].value_counts()
```

```python
df.rename(columns={'variable':'period', 'value':'amount'}, inplace=True)
df.head(10)
```

## Start to Analysis

```python
df.info()
```

```python
df['amount'] = df['amount'].replace('-', np.nan)
df['amount'] = df['amount'].astype(float)
df['amount'].describe()
```

```python
# all cities & gender equal total indexing
df_all = df[(df['시군구별'] == '전국') & (df['gender'] == 'total')]
df_all.head()
```

```python
# we need only year, month amount columns.
df_all = df_all[['year', 'month', 'amount']].copy()
df_all
```

```python
# if you need font set for korean or japanese,
import matplotlib.pyplot as plt
# plt.rc("font", family="Malgun Gothic") for win
# plt.rc("font", family="AppleGothic" ) for mac
df_all.set_index(['year', 'month']).plot(figsize=(15,4))
```

```python
df_all[-36:].set_index(['year', 'month']).plot.bar(figsize=(15,4))
```

```python
# Seaborn
import seaborn as sns
```

```python
plt.figure(figsize = (15,4))
sns.lineplot(data=df_all, x='year', y='amount')
# ci : Confidence Inteval(신뢰구간) 95%
```
