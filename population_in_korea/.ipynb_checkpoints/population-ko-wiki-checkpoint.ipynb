```python
import pandas as pd
```

```python
url = "https://ko.wikipedia.org/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%9D%B8%EA%B5%AC"
```

```python
table = pd.read_html(url) # parse table type on html
len(table)
```

```python
table[0]
```

```python
df = table[1]
df.shape
```

```python
df.head()
```

```python
df.tail()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# win -> sns.set(font="Malgun Gothic")
sns.set(font="AppleGothic")
```

```python
#high-contrast display
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=60)
sns.pointplot(data=df, x="연도 (년)", y="추계인구(명)")
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=60)
sns.lineplot(data=df, x="연도 (년)", y="추계인구(명)")
```
